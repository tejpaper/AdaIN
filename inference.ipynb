```python
from adain import *
```

```python
model = StyleTransferNetwork().eval().to(DEVICE)
model.load_state_dict(torch.load(PATH2MODEL, map_location=DEVICE))
```

```python
stylize(model,
        path2content='images/cornell university.jpg',
        styles=[
            Style('images/woman with a hat.jpg'),
        ],
        save_color=True,
        )
```

```python
stylize(model,
        path2content='images/dancing.jpg',
        styles=[
            Style('images/picasso.jpg', weight=0.5),
            Style('images/lightning.jpg', weight=0.5),
        ],
        save_color=False,
)
```

```python
stylize(model,
        path2content='images/woman.jpeg',
        styles=[
            Style('images/pencil 1.jpg', weight=0.1),
            Style('images/pencil 2.jpg', weight=0.1),
            Style('images/pencil 3.jpg', weight=0.3, color_target=True),
        ],
        save_color=True,
        )
```
